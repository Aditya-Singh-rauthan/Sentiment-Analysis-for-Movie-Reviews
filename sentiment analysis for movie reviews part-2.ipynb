```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('new_reviews.csv')
```

```python
df.head()
```

```python
all_tokens=[]
from nltk.tokenize import word_tokenize
```

```python
for i in range(50000):
    all_tokens+=word_tokenize(df['review'][i])
```

```python
all_tokens
```

```python
words=[]
```

```python
from nltk.corpus import stopwords
```

```python
for token in all_tokens:
    if token in stopwords.words('english'):
        continue
    else:
        words.append(token)
```

```python
wf=pd.DataFrame({'words':words})
```

```python
wf.head()
```

```python
wf.to_csv('tokens_without_stopwords.csv',index=False)
```
